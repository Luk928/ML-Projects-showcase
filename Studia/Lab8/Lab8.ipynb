```python
import os
import tensorflow as tf
import cv2
import numpy as np
```

```python
def load_data(data_directory):
    directories = [d for d in os.listdir(data_directory)
                  if os.path.isdir(os.path.join(data_directory, d))]
    labels = []
    images = []
    for d in directories:
        label_directory = os.path.join(data_directory, d)
        file_names = [os.path.join(label_directory,f)
                     for f in os.listdir(label_directory)
                     if f.endswith('.ppm')]
        for f in file_names:
            # Sckikt Image didn't work so I used opencv
            img = cv2.imread(f)
            im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # OpenCV uses BGR, but pyplot used later uses RGB
            img = np.asarray(im_rgb)
            img = img/255 # Normalizing pixels from 0-255 to 0-1
            images.append(img) 
            labels.append(int(d))
    
    return images, labels
```

```python
root_path = './Data/'
train_data_directory = os.path.join(root_path, 'Training/')
test_data_directory = os.path.join(root_path, 'Testing/')

images, labels = load_data(train_data_directory)
```

```python
images28 = [cv2.resize(image, (28, 28)) for image in images]

traffic_signs = [300, 2250, 3650, 4000]
for i in range(len(traffic_signs)):
    plt.subplot(1,4,i+1)
    plt.axis('off')
    plt.imshow(images28[traffic_signs[i]])
    plt.subplots_adjust(wspace=0.5)
    print("shape: {0}, min: {1}, max: {2}".format(
        images[traffic_signs[i]].shape,
        images[traffic_signs[i]].min(),
        images[traffic_signs[i]].max()))
```

```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray

images_gray = [rgb2gray(image) for image in images28]

traffic_signs = [300, 2250, 3650, 4000]
for i in range(len(traffic_signs)):
    plt.subplot(1,4,i+1)
    plt.axis('off')
    plt.imshow(images_gray[traffic_signs[i]], cmap='gray')
    plt.subplots_adjust(wspace=0.5)
    print("shape: {0}, min: {1}, max: {2}".format(
        images[traffic_signs[i]].shape,
        images[traffic_signs[i]].min(),
        images[traffic_signs[i]].max()))
```

```python
images_rotated = [np.transpose(image) for image in images_gray]

traffic_signs = [300, 2250, 3650, 4000]
for i in range(len(traffic_signs)):
    plt.subplot(1,4,i+1)
    plt.axis('off')
    plt.imshow(images_rotated[traffic_signs[i]], cmap='gray')
    plt.subplots_adjust(wspace=0.5)
    print("shape: {0}, min: {1}, max: {2}".format(
        images[traffic_signs[i]].shape,
        images[traffic_signs[i]].min(),
        images[traffic_signs[i]].max()))
```

```python
import scipy.ndimage as scp
images_blurred = [scp.gaussian_filter(image, 2) for image in images_gray]

traffic_signs = [300, 2250, 3650, 4000]
for i in range(len(traffic_signs)):
    plt.subplot(1,4,i+1)
    plt.axis('off')
    plt.imshow(images_blurred[traffic_signs[i]], cmap='gray')
    plt.subplots_adjust(wspace=0.5)
    print("shape: {0}, min: {1}, max: {2}".format(
        images[traffic_signs[i]].shape,
        images[traffic_signs[i]].min(),
        images[traffic_signs[i]].max()))
    
```
