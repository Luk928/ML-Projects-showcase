Postanowiono przeprowadzić 3 testy korelacji. Testy przeprowadzono dla zestawu danych Iris Dataset i porównywano szerokość działki kielicha oraz długość i szerokość płatka kwiatu do długości działki kielicha. Dla każdego z badanych parametrów przeprowadzono testy Pearsona, Spearmana i Kendalla. W celu lepszej wizualizacji zależności danych od siebie oprócz testów podanych w instrukcji zrobiono również wykresy y = f(x), gdzie y to badany parametr, a x to długość działki kielicha.
</p>

Odnośniki do odpowiednich fragemntów kodu:
1. [Długość płatka kwiatu](#pearson)
2. [Szerokość płatka kwiatu](#spearman)
3. [Szerokość działki kielicha](#kendall)

<p style='text-align:justify;'> 
Dla testu Spearmana zawsze uzyskiwano wartości bardziej oddalone od zera w odpowiednim kierunku (w zależności od tego czy korelacja była pozytywna, czy negatywna), co pozwala sądzić, że test ten ma najmniejsze prawdopodobieństwo nie wykrycia korelacji pomiędzy zmiennymi. Jako, że test Spearmana ma mniej wymaganych założeń niż test Pearsona, to można było się spodziewać tego wyniku. Jednak test Kendalla ma jeszcze mniej założeń, a mimo to wartość jego współczynnika jest jeszcze niższa. W przypadku wykresów 1 i 2 widać, że dla małych wartości korelacja jest znacznie niższa niż dla pozostałych. Można sądzić, że mniejsza ilość założeń w teście Kendalla powoduje, iż obecność fragmentów zaniżonej korelacji wpływa na ten test bardziej niż na pozostałe, co przekłada się na niższą wartość współczynnika korelacji. Test 3 to jedyny, w którym utrzymano negatywną korelację oraz wartość p inną niż 0. Wartość ta oznacza prawdopodobieństwo, że korelacja nie istnieje, a wykrycie jej było przypadkiem. Faktycznie patrząc tylko na wykres punktowy może być ciężko dostrzec korelację, jednak linia trendu wyraźnie wskazuje na jej obecność. Wartości p nie są wysokie, i zapewne są spowodowane głównie dużymi odległościami punktów od linii trendu. Wartości korelacji dla tego przypadku również są niskie, co widać również po nachyleniu linii trendu.
</p><p style='text-align:justify;'> 
Ostatecznie można stwierdzić, że dzięki najmniejszej ilości założeń początkowych test Kendalla oddaje najbardziej rzeczywistą wartość korelacji oraz ma największe prawdopodobieństwo wykrycia tego, że istniejąca korelacja jest dziełem przypadku. Jednak jeżeli komuś bardzo zależy na wykryciu korelacji, to test Spearmana wydaje się być najlepszym rozwiązaniem. Test Pearsona jest swojego rodzaju wypośrodkowaniem pozostałych dwóch testów.
</p>

```python
from scipy.stats import *
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
iris = load_iris()
data = pd.DataFrame(iris.data)
data.columns =['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'] 
data['Class'] = iris.target
data.head()
```

<a id='pearson'></a>

```python
x = data['Sepal_Length']
plt.figure(figsize=(8,8))
plt.scatter(x,data['Petal_Length'])
# Trend line
z = np.polyfit(x, data['Petal_Length'], 1)
p = np.poly1d(z)
plt.plot(x,p(x),"g--")

plt.show()
# Pearson R
r, p = pearsonr(data['Sepal_Length'],data['Petal_Length'])
print("Pearson’s correlation coefficient: %.2f, Two-tailed p-value: %.2f"%(r,p))
# Rang Spearman
r, p = spearmanr(data['Sepal_Length'],data['Petal_Length'])
print("Spearman correlation coefficient: %.2f, The two-sided p-value: %.2f"%(r,p))
# Kendall's Tau
r, p = kendalltau(data['Sepal_Length'],data['Petal_Length'])
print("Tau statistic: %.2f, The two-sided p-value: %.2f"%(r,p))
```

<a id='spearman'></a>

```python
plt.figure(figsize=(8,8))
plt.scatter(x,data['Petal_Width'])
# Trend line
z = np.polyfit(x, data['Petal_Width'], 1)
p = np.poly1d(z)
plt.plot(x,p(x),"g--")


plt.show()
# Pearson R
r, p = pearsonr(data['Sepal_Length'],data['Petal_Width'])
print("Pearson’s correlation coefficient: %.2f, Two-tailed p-value: %.2f"%(r,p))
# Rang Spearman
r, p = spearmanr(data['Sepal_Length'],data['Petal_Width'])
print("Spearman correlation coefficient: %.2f, The two-sided p-value: %.2f"%(r,p))
# Kendall's Tau
r, p = kendalltau(data['Sepal_Length'],data['Petal_Width'])
print("Tau statistic: %.2f, The two-sided p-value: %.2f"%(r,p))
```

<a id='kendall'></a>

```python
plt.figure(figsize=(8,8))
plt.scatter(x,data['Sepal_Width'])

# Trend line
z = np.polyfit(x, data['Sepal_Width'], 1)
p = np.poly1d(z)
plt.plot(x,p(x),"g--")

plt.show()
# Pearson R
r, p = pearsonr(data['Sepal_Length'],data['Sepal_Width'])
print("Pearson’s correlation coefficient: %.2f, Two-tailed p-value: %.2f"%(r,p))
# Rang Spearman
r, p = spearmanr(data['Sepal_Length'],data['Sepal_Width'])
print("Spearman correlation coefficient: %.2f, The two-sided p-value: %.2f"%(r,p))
# Kendall's Tau
r, p = kendalltau(data['Sepal_Length'],data['Sepal_Width'])
print("Tau statistic: %.2f, The two-sided p-value: %.2f"%(r,p))
```
