Let's take a look at the values and see what we can clean. We already have an array of names of columns which contain null values. After checking I've discovered that only the PROPERTYFRACTION contain NaN values disguised as ' '. We'll replace these missing values by simply creating a 'MISSING' category in each column. Doing so will create a dtype conflict in the PROPERTYZIP column. There are a few other columns with conflicting dtypes, but we'll take a closer look at that after replacing the NaN values (in case this process creates new conflicts I've overlooked before).

```python
del test
for col in cols_with_nan:
    nan_idx = data[data[col].isna()].index
    data.at[nan_idx, col]= 'MISSING'
    
nan_idx = data[data['PROPERTYFRACTION']== ' '].index
data.at[nan_idx, 'PROPERTYFRACTION']= 'MISSING'
    
print(data.isna().any())
```

Now let's actually look at changing the dtypes of values so that every column has just one dtype of data in it. Columns with multiple types of data are:
* PROPERTYHOUSENUM
* PROPERTYZIP
* RECORDDATE
* DEEDPAGE
In case of PROPERTYHOUSENUM and PROPERTYZIP the existance of multiple data types is caused by the NaN removing, whoch caused string 'MISSING' values to appear in a column otherwise filled with floats. To fix this we could either drop these rows or return the values back to NaN. When it comes to PROPERTYZIP we could also convert the floats to strings withour information loss (which seems to be the best solution since it forces the model to percieve zip codes as categories, not numerical values). In case of PROPERTYHOUSENUM the numbers reflect property position on a street, so converting them to strings would cause some information loss (although it's arguable whether the information would be of any use to the model). Let's start with analyzing the data.

```python
data[data['PROPERTYHOUSENUM'].apply(lambda x: isinstance(x, str))].shape
```

```python
# Since it's only 18 rows we'll just drop them
nan_idx = data[data['PROPERTYHOUSENUM'].apply(lambda x: isinstance(x, str))].index
data.drop(nan_idx, inplace=True)
# Converting the zipcodes to strings
data['PROPERTYZIP'] = data['PROPERTYZIP'].astype(str)
```

In RECORDDATE some values were missing. It's probably caused by the properties not being registered as of making this dataset.  We could probably drop these rows, but we could also calculate the average time difference between sale date and record date, and then add this average value to the sale dates to impute the missing RECORDDATE values. 
<i>

**WARNING** The date is not missing in only 4230 rows, and it is missing in 297531 rows. This means that the better option might be to drop the entire column, but data cleaning is one of my main goals in this project and there are no other values to try imputation, so I've decided to calculate the missing values anyway despite it being an impractical (and likely wrong) move.</i>

Once we're done with the imputation we'll need to do some more processing necessary to use these features in the model. Using pandas class DatetimeIndex we'll split each of them into 3 new features (year, month, day) and drop the original columns. This will allow us to effectively use LabelEncoder later on (and thus it'll let us efficiently use these features in the training).

```python
# I'm using a function so that I don't have to delete each temporary variable manually to save memory
def date_imputer():    
    test = data[data.RECORDDATE.notnull()]
    sale_dates = np.array(test.SALEDATE.values)
    record_dates = np.array(test.RECORDDATE.values)
    difference = []
    for idx, item in enumerate(record_dates):
        difference.append(record_dates[idx]-sale_dates[idx])
    return np.mean(difference)
mean_date = date_imputer()
sale_dates = data.SALEDATE.values
sale_dates = sale_dates+mean_date
data['TEMP'] = sale_dates # Temporary column to replace missing values
```

```python
data['RECORDDATE'] = np.where(data['RECORDDATE'].isnull, data['TEMP'], data['RECORDDATE']) # Replacement of missing values
data.drop('TEMP', axis = 1, inplace = True)
print(data.RECORDDATE.isnull().any()) # Check to see if there are any null values remaining
```

```python
for col in ['SALEDATE', 'RECORDDATE']:
    data[col+'_YEAR'] = pd.DatetimeIndex(data[col]).year
    data[col+'_MONTH'] = pd.DatetimeIndex(data[col]).month
    data[col+'_DAY'] = pd.DatetimeIndex(data[col]).day
    data.drop(col, axis=1, inplace=True)
data.head()
```

Since everything seems to be fine let's get down to creating the model and trying to predict the property prices. The only thing we could do is scaling the features, but we'll do it in feature preprocessing.

```python
# This code was added later on
nan_idx = data[data.PRICE>700000].index
data.drop(nan_idx, inplace=True) #Deleting the nan positions

# Creating a seperate dataframe with target values
target = data.pop('PRICE')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelBinarizer, LabelEncoder

str_cols = ['PARID', 'PROPERTYFRACTION', 'PROPERTYADDRESSDIR','PROPERTYADDRESSSTREET',
           'PROPERTYADDRESSSUF','PROPERTYADDRESSUNITDESC','PROPERTYUNITNO',
           'PROPERTYCITY','PROPERTYZIP','SCHOOLDESC','MUNIDESC','DEEDBOOK',
           'SALECODE','SALEDESC','INSTRTYP','INSTRTYPDESC']

clfs = {c:LabelEncoder() for c in str_cols}

for col, clf in clfs.items():
    data[col] = clfs[col].fit_transform(data[col])
    
# Feature scaling
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data)
data = pd.DataFrame(data).astype('float32')

# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=123)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
# grid_params = {
#     'n_estimators': [190,200,210],
#     'min_samples_split': [2],
#     'min_samples_leaf':[1]
# }
# model = RandomForestRegressor(n_estimators=200)
# gs = GridSearchCV(model, grid_params, verbose = 1, n_jobs=-1, cv=3)
# gs_results = gs.fit(X_train, y_train)
# print(gs_results.best_params_)
# print(gs_results.best_score_)
```

```python
model = RandomForestRegressor(n_estimators=200, n_jobs=-1)
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

Tested algorithms: K-nearest neighbors, support vector machine Reressor, linear regression and random forest.
The best one was random forest, and the worst one was SVR (it had negative R2 coefficient).
The best random forest model (obtained using grid search cross-validation) was {'min_samples_leaf': 1, 'min_samples_split': 2, 'n_estimators': 200}. For all samples it achieved 0.39 R2 coefficient. Looking at the price histogram it's quite easy to see, that there are multiple outliers with very high price. In many datasets dealing with properties there is a feature containing information about the property area, but in this dataset there is no such feature. It's a shame, beacuse it's a very valuable and important parameter that is highly correlated with price. Because there is no such feature in this dataset, it means that there is some threshold price value, from which the model no longer has any way of accurately predicting the price, because no other features change with it. To test my thesis I decided to drop values with high prices. First I decided to drop everything with price higher than million. The R2 coefficient jumped from 0.39 to 0.68, so I decided to drop all properties with price over 700'000. This caused the R2 coefficient to reach 0.96, at which point I decided to stop dropping the most valuable properties. This model also achieved 0.72 on the test set, so the model is overfitting. Sadly, lack of data about area ot other feature highly correlated to price makes it impossible to have the model trained well with no over- or underfitting.
